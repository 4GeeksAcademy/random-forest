```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Gráfico de línea simple:
plt.title('Aplicación de la función logarítmica')
plt.xlabel("Valores originales")
plt.ylabel("Valores transformados")
x = np.linspace(1, 50, 100)
y = np.log(x)
plt.plot(x, y, color='blue', label='Valores', linestyle='-', marker='*')
plt.legend()
plt.show()
```

```python
# Scatter plot:
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
plt.scatter(x, y, color='blue', label='Valores', marker='*')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión')
plt.show()
```

```python
# Histograma:
data = np.random.randn(1000)
plt.hist(data, bins=20)
plt.title("Histograma")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Gráficos de barras:
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 15, 30]
plt.bar(categories, values)
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
plt.show()
```

```python
# Gráfico de pastel:
labels = ['A', 'B', 'C', 'D']
sizes = np.random.randint(low=1, high=20, size=4)
plt.pie(sizes, labels=labels)
plt.title("Gráfico de pastel")
plt.legend()
plt.show()
```

```python
# Subplots:
fig, ax = plt.subplots(2, 2, figsize=(13, 8)) # 2 filas, 2 columnas

# Subplot 1 - Gráfico de línea
ax[0, 0].plot([1, 2, 3, 4], [1, 4, 2, 3])
ax[0, 0].set_title('Gráfico de Línea')
ax[0, 0].set_xlabel('Eje X')
ax[0, 0].set_ylabel('Eje Y')
ax[0, 0].legend(['Serie 1'])

# Subplot 2 - Gráfico de barras
ax[0, 1].bar(['A', 'B', 'C', 'D'],[1, 2, 3, 4])
ax[0, 1].set_title('Gráfico de Barras')
ax[0, 1].set_xlabel('Categorías')
ax[0, 1].set_ylabel('Valores')

# Subplot 3 - Gráfico de dispersión
ax[1, 0].scatter([1, 2, 3, 4], [1, 4, 2, 3])
ax[1, 0].set_title('Gráfico de Dispersión')
ax[1, 0].set_xlabel('Eje X')
ax[1, 0].set_ylabel('Eje Y')

# Subplot 4 - Gráfico de pastel
ax[1, 1].pie([10, 20, 30, 40], labels=['A', 'B', 'C', 'D'])
ax[1, 1].set_title('Gráfico de Pastel')
ax[1, 1].legend()

plt.tight_layout()  # Ajustar el espacio entre subplots automáticamente
plt.show()
```

```python
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_df = pd.read_csv(url)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(titanic_df['Age'].dropna(), bins=20, edgecolor='k')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades de los Pasajeros del Titanic')
plt.show()
```

```python
class_counts = titanic_df['Pclass'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(class_counts.index, class_counts.values, tick_label=['1ra Clase', '2da Clase', '3ra Clase'])
plt.xlabel('Clase')
plt.ylabel('Cantidad de Pasajeros')
plt.title('Cantidad de Pasajeros por Clase en el Titanic')
plt.show()
```

```python
port_counts = titanic_df['Sex'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(port_counts, labels=port_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Para que el gráfico sea un círculo
plt.title('Distribución de Puertos de Embarque en el Titanic')
plt.show()
```
