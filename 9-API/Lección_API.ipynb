```python
# Possible intervals:'1m','3m','5m','15m','30m','1h','2h','4h','6h','8h','12h','1d'
data = requests.get('https://api.binance.com/api/v3/klines', params={'symbol':'ETHUSDT', 'interval':'1d','limit': 7}).json()
data[0]
```

```python
import numpy as np
import pandas as pd
from datetime import datetime

# If succesful we create a dataframe with the data retrieved and store it in our dictionary
df = pd.DataFrame(data=np.array(data)[:,0:7],columns=['Open_time','Open','High','Low','Close','Volume','Close_time'])
# Chane UNIX time to Datetime format
df['Open_time'] = df['Open_time'].apply(lambda x: datetime.utcfromtimestamp(int(x)/1000).strftime('%y-%m-%d %H:%M'))
df['Close_time'] = df['Close_time'].apply(lambda x: datetime.utcfromtimestamp(int(x)/1000).strftime('%y-%m-%d %H:%M'))
```

# Ejercicio 4: Obtener redes de bicicletas disponibles en citybik y obtener informaci√≥n de las estaciones de una de las redes

```python
def get_networks():
    url = 'https://api.citybik.es/v2/networks'

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        return data['networks']
    else:
        print(f"Error al obtener los datos. C√≥digo de estado: {response.status_code}")
        return None

redes_bicis = get_networks()
```

```python
redes_bicis
```

```python
def get_stations(network_id):
    url = f'https://api.citybik.es/v2/networks/{network_id}'
    response = requests.get(url)
    print(response.status_code)
    if response.status_code == 200:
        data = response.json()
        if 'stations' in data['network'] and isinstance(data['network']['stations'], list):
            return data['network']['stations']
        else:
            print(f"No se encontraron estaciones para la red '{network_id}'.")
            return []
    else:
        print(f"Error al obtener los datos. C√≥digo de estado: {response.status_code}")
        return None

# Ejemplo de uso
network_id = 'sevici'
estaciones = get_stations(network_id)

if estaciones:
    print(f"Estaciones disponibles para la red '{network_id}':")
    for estacion in estaciones:
        print(f"ID: {estacion['id']}, Nombre: {estacion['name']}, Latitud: {estacion['latitude']}, Longitud: {estacion['longitude']}")
else:
    print("No se pudo obtener la informaci√≥n de las estaciones.")
```
